```python
# draw a star with xkcd matplotlib
import matplotlib.pyplot as plt
import numpy as np
import matplotlib.path as mpath

# only outer edges
star_path = mpath.Path.unit_regular_star(5, innerCircle=0.4)
vertices = star_path.vertices  # Returns a (11, 2) array of coordinates
lines = [
    [vertices[i], vertices[i + 1]] for i in range(len(vertices) - 1)
]

# one continuous line instead
lines = [vertices]

# turn off xkcd
plt.xkcd(False)

plt.figure(figsize=(1, 1))
plt.xkcd(scale=1)
for line in lines:
    x, y = zip(*line)
    plt.plot(x, y, color='blue', linewidth=5)
    # fill the star
    plt.fill(x, y, color='blue', alpha=1)

plt.axis('equal')
plt.axis('off')
plt.legend(frameon=False)

lim_padding = 1.0
xlim, ylim = plt.xlim(), plt.ylim()
plt.xlim(xlim[0]*lim_padding, xlim[1]*lim_padding)
plt.ylim(ylim[0]*lim_padding, ylim[1]*lim_padding)

plt.gca().axis('off')
plt.gca().set_aspect('equal')

# 2. Make the background of both the figure and the axes transparent
fig = plt.gcf()
fig.patch.set_alpha(0.0)
plt.gca().patch.set_alpha(0.0)

plt.gca().set_facecolor('none')
plt.gca().set_frame_on(False)


# remove background
# do not export with a white background

# export as svg
plt.savefig('assets/images/indicators/star.svg', format='svg', transparent=True)
```

```python
import matplotlib.pyplot as plt
import numpy as np
import os

# Create directory if it doesn't exist
os.makedirs('assets/images/indicators/', exist_ok=True)

def draw_and_save_arrow(direction='right', filename='arrow.svg'):
    plt.figure(figsize=(1, 1))
    # scale=1 for that signature hand-drawn wobble
    plt.xkcd(scale=1, length=100, randomness=2)
    
    if direction == 'right':
        # Vertices for a right-pointing arrow
        vertices = np.array([
            [0.1, 0.4], [0.5, 0.4], [0.5, 0.2], [0.9, 0.5], 
            [0.5, 0.8], [0.5, 0.6], [0.1, 0.6], [0.1, 0.4]
        ])
    else:
        # Vertices for a left-pointing arrow
        vertices = np.array([
            [0.9, 0.4], [0.5, 0.4], [0.5, 0.2], [0.1, 0.5], 
            [0.5, 0.8], [0.5, 0.6], [0.9, 0.6], [0.9, 0.4]
        ])

    x, y = zip(*vertices)
    
    # Plot outline and fill
    # plt.plot(x, y, color='black', linewidth=3)
    plt.fill(x, y, color='black', alpha=1)

    # Styling the plot area
    plt.axis('equal')
    plt.axis('off')
    
    # Ensure background transparency
    fig = plt.gcf()
    fig.patch.set_alpha(0.0)
    plt.gca().patch.set_alpha(0.0)
    plt.gca().set_facecolor('none')
    
    # Set limits to prevent clipping
    plt.xlim(-0.1, 1.1)
    plt.ylim(-0.1, 1.1)

    # Export
    plt.savefig(f'assets/images/indicators/{filename}', format='svg', transparent=True)
    plt.show()
    plt.close()

# Execute for both directions
draw_and_save_arrow(direction='left', filename='arrow_left.svg')
draw_and_save_arrow(direction='right', filename='arrow_right.svg')

print("Arrows saved to assets/images/indicators/")
```

```python
import matplotlib.pyplot as plt
import numpy as np
import os

# Create directory
os.makedirs('assets/images/indicators/', exist_ok=True)

def draw_custom_arrow(direction='right', style='block', stem_len=0.4, stem_wd=0.2, 
                      head_wd=0.6, head_len=0.3, gap=0.05, lw=3, color='black', 
                      filename='arrow.svg'):
    """
    stem_len: Length of the tail section
    stem_wd:  Thickness of the tail section
    head_wd:  Vertical width (height) of the arrow head
    head_len: Horizontal length of the arrow head
    gap:      Spacing between the stem and the head
    lw:       Line width for the stroke
    """
    plt.figure(figsize=(1, 1))
    plt.xkcd(scale=1, length=100, randomness=1.5)
    
    mid = 0.5
    # Calculate X coordinates based on direction
    if direction == 'right':
        # Stem starts at 0.1, goes for stem_len
        s_start, s_end = 0.1, 0.1 + stem_len
        # Head starts after the stem + gap
        h_start, h_tip = s_end + gap, s_end + gap + head_len
    else:
        # Stem starts at 0.9, goes left
        s_start, s_end = 0.9, 0.9 - stem_len
        # Head starts after stem + gap (moving left)
        h_start, h_tip = s_end - gap, s_end - gap - head_len

    # Y coordinates
    s_top, s_bot = mid + (stem_wd/2), mid - (stem_wd/2)
    h_top, h_bot = mid + (head_wd/2), mid - (head_wd/2)

    if style == 'block':
        # Draw Stem
        stem_v = np.array([[s_start, s_bot], [s_end, s_bot], [s_end, s_top], [s_start, s_top], [s_start, s_bot]])
        plt.fill(stem_v[:,0], stem_v[:,1], color=color)
        plt.plot(stem_v[:,0], stem_v[:,1], color=color, linewidth=lw)
        
        # Draw Head
        head_v = np.array([[h_start, h_bot], [h_tip, mid], [h_start, h_top], [h_start, h_bot]])
        plt.fill(head_v[:,0], head_v[:,1], color=color)
        plt.plot(head_v[:,0], head_v[:,1], color=color, linewidth=lw)
        
    elif style == 'line':
        # Draw Shaft
        plt.plot([s_start, s_end], [mid, mid], color=color, linewidth=lw)#, capstyle='round')
        # Draw Head (v-shape)
        plt.plot([h_start, h_tip, h_start], [h_top, mid, h_bot], color=color, linewidth=lw)#, capstyle='round')

    plt.axis('equal')
    plt.axis('off')
    # Set limits with a bit of padding to ensure the xkcd "wobble" isn't cut off
    plt.xlim(-0.05, 1.05)
    plt.ylim(-0.05, 1.05)
    
    fig = plt.gcf()
    fig.patch.set_alpha(0.0)
    plt.gca().patch.set_alpha(0.0)
    
    plt.savefig(f'assets/images/indicators/{filename}', format='svg', transparent=True)
    plt.show()
    plt.close()

# --- Examples ---

# 1. Thick Block Arrow with a significant gap
# draw_custom_arrow(direction='right', style='block', stem_len=0.3, gap=0.1, lw=2, filename='gap_block.svg')

# 2. Thin Line Arrow with no gap and heavy line width
draw_custom_arrow(direction='left', style='line', stem_len=0.8, gap=-0.2, head_wd=0.5, lw=5, filename='arrow_left.svg')
draw_custom_arrow(direction='right', style='line', stem_len=0.8, gap=-0.2, head_wd=0.5, lw=5, filename='arrow_right.svg')

# 3. Modern Minimalist (Short stem, large head)
# draw_custom_arrow(direction='right', style='block', stem_len=0.1, stem_wd=0.05, head_wd=0.8, head_len=0.4, gap=0.05, filename='minimal.svg')
```
